```python
log_file = 'exp_b0_1.log'
# load the gpu memory usage from the log file
with open(log_file, 'r') as f:
    lines = f.readlines()
    gpu_memory = []
    for line in lines:
        if 'GPU 0 - Memory Used: ' in line:
            gpu_memory.append(int(line.split('GPU 0 - Memory Used: ')[1].split('/')[0].split('MB')[0]))
    gpu_memory = np.array(gpu_memory)
gpu_memory
```

```python
plt.plot(gpu_memory[:100]/1024)
```
